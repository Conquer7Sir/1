```python
年初固定资产净值 = 100000       # 第一年年初固定资产净值 = 固定资产原值
result = []                     # 创建空列表存储年折旧额

for i in range(8):              
    年折旧额 = round(年初固定资产净值 * (2/10),2)       # 年折旧率 = 2/10
    result.append(年折旧额)
    年初固定资产净值 -= 年折旧额
else:
    年折旧额 = round((年初固定资产净值-100000*0.05)/2,2)
    result.append(年折旧额)
    result.append(round(年初固定资产净值-100000*0.05-年折旧额,2))
print("每年折旧额依次为：",result)
```

```python
def 折旧额计算(原值,折旧年限,残值率):

    年初固定资产净值 = 原值       # 第一年年初固定资产净值 = 固定资产原值
    result = []                     # 创建空列表存储年折旧额

    for i in range(折旧年限):
        if i < 折旧年限 -2:
            年折旧额 = round(年初固定资产净值 * (2/折旧年限),2)       # 年折旧率 = 2/10
            result.append(年折旧额)
            年初固定资产净值 = round(年初固定资产净值 - 年折旧额,2)
        else:
            年折旧额 = round((年初固定资产净值-原值*残值率)/(折旧年限-i),2)
            result.append(年折旧额)
        年初固定资产净值 = round(年初固定资产净值 - 年折旧额,2)
    print("每年折旧额依次为：",result)
    return result
#测试函数：
result = 折旧额计算(100000,10,0.05)
print("折旧结果：",result)
```

```python
loc索引器具体方式:

①选择单行/多行数据: df.loc['行']、df.loc[['行1','行2']]；

②选择多行多列数据： df.loc[['行1','行2'],['列1','列2']]，通过两个列表选取行列组合；

③loc布尔选择: df.loc[(df['列']>条件)],按条件选取单列（多列）满足一定条件的行；

④loc切片(选择连续的多行多列): df.loc['行1':'行2','列1':'列2'],通过**切片选取连续**的行列组合。
老师说：# （a） 选择单行/多行数据：df.loc[’行’]、df.loc[[’行1’,’行2’]]
df.loc[[2,4,6]]   # 注意：多行以列表[2,4,6]形式呈现
老师说：# （b） 选择多行多列数据： df. loc[['行1','行2']，['列1','列2']]，通过两个列表选取行列组合
df.loc[[2,4,6,8],['年','月','流动负债','所有者权益']]
老师说：# （c） loc布尔选择: df.loc[(df['列']>条件)],按条件选取单列（多列）满足一定条件的行
df.loc[df['流动负债']>100000,['年','月','流动资产','流动负债','所有者权益']]
老师说：# （d） loc切片（选择连续多行多列） ：df.loc[’行1':'行2','列1':'列2']，通过切片选取
df.loc[1:10, '流动资产':'非流动负债']
```

```python
# 引入pandas
import pandas as pd
# 读取利润表项目
df = pd.read_excel('E:\data.xlsx',sheet_name = 1,converters= {'年':str,'月':str})
df
#查看分组,并展示结果
df.groupby('年').get_group ('2021')
# 要求:使用groupby函数按“年”对data. xlsx中的利润表项目进行分组求和。
#按年分组求和,并展示结果
df.groupby('年').sum()
df.groupby('年').sum().iloc[:,-3:]
# 2.1 对一列使用多种聚合
# 要求：使用groupby函数按“年”对data.xlsx中的利润表项目进行分组，并统计各组“营业收入”列的平均值、合计值
df1 = df.groupby('年').agg({'营业收入': ['mean','sum']}).round (2)
df1
# 2.2 对每列使用相同的多种聚合
# 要求：使用groupby函数按“年”对data. xlsx中的利润表项目进行分组，并统计各组每列的平均值、合计值、最小值、最大值。
df2 = df.groupby('年').agg(['mean','sum','min']).round (2).iloc[:,-6:]
df2
# 2.3 对选定列使用不同的聚合
# 要求：使用groupby函数按“年”对data.xlsx中的利润表项目进行分组，并统计各组“营业收入”的最大值、“营业成本”的最小值、“净利润”的合计值
df3 = df.groupby('年').agg({'营业收入':[max,min],'营业成本':min,'净利润':[max,min]})                            
df3
```

```python
# 引入pandas
import pandas as pd
# 读取利润表项目
df = pd.read_excel('E:\data.xlsx',sheet_name = 1,converters= {'年':str,'月':str})
df
#创建一个月份与季度对照的字典monthToQuarter
monthToQuarter = {'1':'第1季','2':'第1季','3':'第1季',
                  '4':'第2季','5':'第2季','6':'第2季',
                  '7':'第3季','8':'第3季','9':'第3季',
                  '10':'第4季','11':'第4季','12':'第4季'}

# 新增‘季度’列
df['季度'] = df['月'].map(monthToQuarter)

# df.head()
df
# 数据透视表
df1 = df.pivot_table(index='季度',columns='年',values=['营业收入','营业成本'],aggfunc='sum')
df1
df2 = df.pivot_table(index=['季度','月'],columns='年',values=['营业收入','营业成本'],aggfunc='sum')
df2
df3 = df1.stack()
df3
# df4 = df3.unstack(level=-1)
df4 = df3.unstack()
df4
df4 = df2.unstack(level=0)
df4
```
